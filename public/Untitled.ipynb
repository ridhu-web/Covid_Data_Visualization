```python
import csv

```

```python
encoding = 'ISO-8859-1'
```

```python
# Function to convert CSV to JavaScript Map
def csv_to_js_map(csv_file_path, key_column,check_column, value_column):
    js_map = "const myMap = new Map(["

    with open(csv_file_path, mode='r', encoding=encoding, errors='replace') as file:
        reader = csv.DictReader(file)
        reader = csv.DictReader(file)
        for row in reader:
            print(row)
            key = row[key_column]
            if row[check_column] == "N_POP_CHG_2021":
                value = row[value_column]
                js_map += f'{key}:{value},'

    js_map = js_map.strip(', ') + "]);"

    return js_map
```

```python
# Example usage
csv_file_path = 'PopulationEstimates.csv'
```

```python
key_column = 'FIPStxt'
check_column = 'Attribute'
value_column = 'Value'
```

```python
javascript_map = csv_to_js_map(csv_file_path, key_column,check_column, value_column)


```
